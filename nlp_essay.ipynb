Let's move to a more complicated example!

```python
class Triangle:
    def __init__(self, base, height, side_a, side_b, side_c):
        self.base = base
        self.height = height
        self.a = side_a
        self.b = side_b
        self.c = side_c

    def area(self):
        return 0.5 * self.base * self.height

    def perimeter(self):
        return self.a + self.b + self.c
```

```python
#Thanks to this code, we can now calculate the area of a triangle with the base = 6, height = 4, side a = 5, side b = 6, and side c = 7.
triangle1 = Triangle(base=6, height=4, side_a=5, side_b=6, side_c=7)

print("Area:", triangle1.area())
print("Perimeter:", triangle1.perimeter())
```

Now let's use Python and NLTK to perform a basic linguistic analysis on some text data scraped from the Internet.

Link to the website: https://csi.amu.edu.pl/en/blog-csi/nlp-what-is-it-and-what-can-it-be-used-for

```python
import requests
```

```python
page = requests.get("https://csi.amu.edu.pl/en/blog-csi/nlp-what-is-it-and-what-can-it-be-used-for")
# print(page.content)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(page.content, 'html.parser')
# soup
```

```python
paras = soup.find_all('p')
# print(paras)
```

```python
only_text = []

for el in paras:
    if len(el.getText().strip()) > 0:
        only_text.append(el.getText())

print(only_text)
```

```python
import nltk
```

```python
nltk.download('punkt_tab')
tuples = []

for sentence in only_text:
  tokenized = nltk.word_tokenize(sentence)
  print(tokenized)
```

```python
nltk.download('averaged_perceptron_tagger_eng')
tuples = []

for sentence in only_text:
  tokenized = nltk.word_tokenize(sentence)
  tagged = nltk.pos_tag(tokenized)
  for item in tagged:
    tuples.append(item)

print(tuples)
```

```python
counter_dict = {}
for item in tuples:
    print(item)
    if item[1] not in counter_dict:
        counter_dict[item[1]] = 1
    else:
        counter_dict[item[1]] += 1

print(counter_dict)
```

```python
import matplotlib.pyplot as plt
```

```python
import matplotlib.pyplot as plt

keys = []
values = []
for el in counter_dict:
    keys.append(el)
    values.append(counter_dict[el])

print(keys)
print(values)
```

```python
plt.bar(keys[0:10], values[0:10], color = "tab:orange")
```

```python
plt.pie(values[0:10], labels=keys[0:10])
```

## The most common words

Now I will perform a word frequency analysis on the article by writing code that counts and identifies the most common words present in the text.

```python
word_counter = {}
for item in tuples:
    word = item[0]
    if word not in word_counter:
        word_counter[word] = 1
    else:
        word_counter[word] += 1

print(word_counter)
```

```python
import string

word_counter = {}
for item in tuples:
    word = item[0]

    if word not in string.punctuation and word.isalpha():
        if word not in word_counter:
            word_counter[word] = 1
        else:
            word_counter[word] += 1

print(word_counter)
```

```python
!pip install matplotlib
```

```python
import matplotlib.pyplot as plt

top_words = sorted(word_counter.items(), key=lambda x: x[1], reverse=True)[:10]

words = [item[0] for item in top_words]
counts = [item[1] for item in top_words]

plt.figure(figsize=(10,6))
plt.bar(words, counts, color='skyblue')
plt.title('Top 10 Most Common Words')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

The most common words are: the, of, a, to, in, is, by, for, on, and.

However, those are so-called **stopwords** and they have little meaning. Let's see if we can do something about it!

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))
```

```python
word_counter = {}
for item in tuples:
    word = item[0].lower()
    if word.isalpha() and word not in stop_words:
        if word not in word_counter:
            word_counter[word] = 1
        else:
            word_counter[word] += 1

print(word_counter)
```

```python
import matplotlib.pyplot as plt

top_words = sorted(word_counter.items(), key=lambda x: x[1], reverse=True)[:10]

words = [item[0] for item in top_words]
counts = [item[1] for item in top_words]

plt.figure(figsize=(12,6))
plt.bar(words, counts, color='skyblue')
plt.title('Top 10 Most Common Words (excluding stopwords)')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.show()
```

### **Sentiment Analysis**

Here is my sentiment analysis of the fragment about the sentiment analysis from the article _NLP – what is it and what can it be used for?_

```python
!pip install textblob
!python -m textblob.download_corpora
```

```python
from textblob import TextBlob

text = "Sentiment analysis is a type of supervised text classification that serves to categorise the opinions expressed in texts. In most cases a division is made into three classes: positive, negative and neutral. In some applications, sentiment analysis is applied to specific aspects of the opinions. For instance, in the case of hotel reviews, the opinions contained in the analysed texts may be categorised with respect to such aspects as price, location, service quality, and quality of meals. For example, in a master’s degree thesis by Kinga Kramer titled Sentiment Analysis of Students’ Comments on Lectures (2018), an experiment is described involving sentiment analysis of comments made by students concerning lectures given by the author of this blog."

blob = TextBlob(text)

print("Polarity:", blob.sentiment.polarity)
print("Subjectivity:", blob.sentiment.subjectivity)
```

## How to read the sentiment score from TextBlob?
The polarity score ranges from -1 to 1, where -1 indicates very negative words such as "disgusting" or "awful" and 1 represents very positive words like "excellent" or "best."

The subjectivity score ranges from 0 to 1. A value closer to 1 means the text contains more personal opinions rather than objective facts.

You can read the whole explanation here: https://www.analyticsvidhya.com/blog/2021/10/sentiment-analysis-with-textblob-and-vader/

**Polarity:** 0.085

The polarity score indicates that the text is **mostly neutral**, with a slight leaning towards a positive tone. This suggests that the language used is generally objective but carries a subtle positive sentiment.

The subjectivity score, which is closer to the middle of the scale, implies that the author primarily focuses on presenting **factual information** while keeping personal opinions to a minimum. This balance is typical for informative or explanatory texts where clarity and objectivity are important.
